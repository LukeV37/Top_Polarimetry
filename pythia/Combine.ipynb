```python
import uproot
import awkward as ak
```

```python
with uproot.open("dataset.root:fastjet") as f:
    jet_pt = f['jet_pt'].array()
    jet_eta = f['jet_eta'].array()
    jet_phi = f['jet_phi'].array()
    jet_m = f['jet_m'].array()
    trk_pt = f['trk_pt'].array()
    trk_eta = f['trk_eta'].array()
    trk_phi = f['trk_phi'].array()
    trk_m = f['trk_m'].array()
    trk_q = f['trk_q'].array()
    trk_d0 = f['trk_d0'].array()
    trk_z0 = f['trk_z0'].array()
    trk_origin =  f['trk_origin'].array()
    
with uproot.open("../madgraph/labels.root:labels") as f:
    costheta = f['costheta'].array()

no_jets = ak.num(jet_pt)!=0

jet_pt = jet_pt[no_jets]
jet_eta = jet_eta[no_jets]
jet_phi = jet_phi[no_jets]
jet_m = jet_m[no_jets]

trk_pt = trk_pt[no_jets]
trk_eta = trk_eta[no_jets]
trk_phi = trk_phi[no_jets]
trk_m = trk_m[no_jets]
trk_q = trk_q[no_jets]
trk_d0 = trk_d0[no_jets]
trk_z0 = trk_z0[no_jets]
trk_origin =  trk_origin[no_jets]

costheta = costheta[no_jets]

trk_pt = ak.flatten(trk_pt,axis=2)
trk_eta = ak.flatten(trk_eta,axis=2)
trk_phi = ak.flatten(trk_phi,axis=2)
trk_m = ak.flatten(trk_m,axis=2)
trk_q = ak.flatten(trk_q,axis=2)
trk_d0 = ak.flatten(trk_d0,axis=2)
trk_z0 = ak.flatten(trk_z0,axis=2)
trk_origin =  ak.flatten(trk_origin,axis=2)
```

```python
with uproot.recreate("data.root") as f:
    f['data'] = {
                "jet": ak.zip({"pt": jet_pt,
                                   "eta": jet_eta,
                                   "phi": jet_phi,
                                   "m" : jet_m,
                                }),
                "trk": ak.zip({"pt": trk_pt,
                                   "eta": trk_eta,
                                   "phi": trk_phi,
                                   "m" : trk_m,
                                   "q": trk_q,
                                   "d0": trk_d0,
                                   "z0": trk_z0,
                                   "origin" : trk_origin,
                                }),
                "costheta": costheta,
                }
    f['data'].show()
```
