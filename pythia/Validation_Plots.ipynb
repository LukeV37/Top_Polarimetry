```python
import uproot
import awkward as ak
import matplotlib.pyplot as plt
```

```python
with uproot.open('dataset.root:fastjet') as f:
    #print(f.keys())
    jet_pt = f['jet_pt'].array()
    jet_eta = f['jet_eta'].array()
    jet_phi = f['jet_phi'].array()
    jet_m = f['jet_m'].array()
    trk_pt = f['trk_pt'].array()
    trk_eta = f['trk_eta'].array()
    trk_phi = f['trk_phi'].array()
    trk_m = f['trk_m'].array()
    trk_q = f['trk_q'].array()
    trk_d0 = f['trk_d0'].array()
    trk_z0 = f['trk_z0'].array()
    trk_origin =  f['trk_origin'].array()
```

```python
def plot_hist(data, Title, xlabel, Range, bins, color, log=False):
    counts, bins, _ = plt.hist(ak.ravel(data), bins=bins, range=Range,histtype='step',density=False,color=color)
    plt.xlabel(xlabel,loc="right")
    plt.ylabel("Count",loc='top')
    plt.title(Title)
    plt.ylim([1,ak.max(counts)*2])
    #plt.legend()
    if log:
        plt.yscale('log')
    plt.show()
```

```python
plot_hist(jet_pt,"Jet pT","GeV",(0,1000),100,'r',True)
plot_hist(jet_eta,"Jet Eta","Eta",(-6,6),100,'r',True)
plot_hist(jet_phi,"Jet Phi","Phi",(0,6.28),100,'r',True)
plot_hist(jet_m,"Jet Mass","GeV",(0,1000),100,'r',True)
```

```python
plot_hist(trk_pt,"Track pT","GeV",(0,600),100,'b',True)
plot_hist(trk_eta,"Track Eta","Eta",(-6,6),100,'b',True)
plot_hist(trk_phi,"Track Phi","Phi",(-4,4),100,'b',True)
plot_hist(trk_m,"Track Mass","GeV",(0,2),100,'b',True)
plot_hist(trk_q,"Track Charge","Charge",(-2,2),100,'b',True)
plot_hist(trk_d0,"Track d0","mm",(-100,100),100,'b',True)
plot_hist(trk_z0,"Track z0","mm",(-100,100),100,'b',True)
plot_hist(trk_origin,"Track Origin","GeV",(-25,25),100,'b',True)
```
